```python
import pandas as pd

df = pd.read_csv("DataScientist.csv")
df.head()
```

```python
df.info()
df.columns
```

```python
df = df.dropna(subset=["Job Description"]).copy()
df['Job Description'] = df['Job Description'].astype(str).str.lower()


skills = [
    'python','java','sql','excel','tableau','power bi','machine learning',
    'deep learning','r','c++','cloud','aws','azure','docker','hadoop','spark',
    'sas','statistics','nlp','pytorch','tensorflow'
]


def extract_skills(text):
    return [s for s in skills if s in text]

df['Extracted Skills'] = df['Job Description'].apply(extract_skills)
df[['Job Title','Extracted Skills']].head()
```

```python
from collections import Counter

all_skills = sum(df['Extracted Skills'], [])
skill_counts = Counter(all_skills)

skill_df = pd.DataFrame(skill_counts.items(), columns=['Skill','Count']) \
             .sort_values(by='Count', ascending=False)

print(skill_df.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


plt.figure(figsize=(10,6))
sns.barplot(x='Count', y='Skill', data=skill_df.head(10), palette="viridis")
plt.title("Top 10 In-Demand Skills", fontsize=14)
plt.show()


text = " ".join(all_skills)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Skill Demand Word Cloud", fontsize=16)
plt.show()
```

```python

region_skills = df.explode('Extracted Skills')


if 'Location' in df.columns:
    region_summary = region_skills.groupby(['Location','Extracted Skills']).size().reset_index(name='Count')

    print("📊 Sample Regional Skill Demand:")
    print(region_summary.head(15))


    region_name = "New York"
    top_region_skills = region_summary[region_summary['Location'].str.contains(region_name, case=False, na=False)]
    top_region_skills = top_region_skills.sort_values(by="Count", ascending=False).head(10)

    plt.figure(figsize=(10,6))
    sns.barplot(x='Count', y='Extracted Skills', data=top_region_skills, palette="mako")
    plt.title(f"Top Skills in {region_name}", fontsize=14)
    plt.show()
```

```python
if 'Industry' in df.columns:
    print("\n✅ Performing Industry Skill Analysis...")


    if 'Extracted Skills' not in df.columns:
        df['Extracted Skills'] = df['Job Description'].apply(extract_skills)

    industry_skills = df.explode('Extracted Skills')

    industry_summary = industry_skills.groupby(['Industry','Extracted Skills']).size().reset_index(name='Count')
    print("\n📊 Sample Industry Skill Demand:")
    print(industry_summary.head(15))


    industry_name = "tech"
    top_industry_skills = industry_summary[
        industry_summary['Industry'].str.contains(industry_name, case=False, na=False)
    ]
    top_industry_skills = top_industry_skills.sort_values(by="Count", ascending=False).head(10)

    if not top_industry_skills.empty:
        plt.figure(figsize=(10,6))
        sns.barplot(x='Count', y='Extracted Skills', data=top_industry_skills, palette="rocket")
        plt.title(f"Top Skills in industries containing '{industry_name}'", fontsize=14)
        plt.show()
    else:
        print(f"⚠️ No matching industry found for '{industry_name}'")
```
